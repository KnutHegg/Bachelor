```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import datetime 
```

# Stationarity


```python
plt.figure(figsize= (12, 10))

    


x = np.arange(0,10* np.pi, 0.01)
y = np.sin(x)

plt.plot(x, y)
plt.axhline(y = 0, color = "black")
plt.ylabel('y')



# Create scatter plot of Positive Cases
#

plt.plot_date(x, y, color = 'blue')
plt.show()
```

```python
plt.figure(figsize= (12, 10))

    


x = np.arange(0,10* np.pi, 0.01)
y = np.sin(x)+x/15

print(x)

sns.lineplot(x, y)
plt.axhline(y = 0, color = "black")
plt.ylabel('$y$')



# Create scatter plot of Positive Cases
#

plt.plot_date(x, y, color = 'blue')
plt.show()
```

```python
plt.figure(figsize= (12, 10))

    


x = np.arange(0,30* np.pi, 0.01)
y = np.exp(-0.04*x)*np.sin(0.5*x)

print(x)

sns.lineplot(x, y)
plt.axhline(y = 0, color = "black")
plt.ylabel('$y$')



# Create scatter plot of Positive Cases
#

plt.plot_date(x,y, color = 'blue')
plt.show()
```
